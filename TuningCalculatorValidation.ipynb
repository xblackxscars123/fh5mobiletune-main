```python
def validate_tune_settings(tune, car_specs, tune_type, pi_class):
    """
    Validate a generated tune against Forza Horizon 5 game limits.
    Returns a validation report with pass/fail status for each parameter.
    """
    validation_results = {
        "car": car_specs.get("name", "Unknown"),
        "tune_type": tune_type,
        "pi_class": pi_class,
        "passed": True,
        "errors": [],
        "warnings": [],
        "details": {}
    }
    
    # Tire pressure validation
    if tune.get("tirePressureFront"):
        front_psi = tune["tirePressureFront"]
        rear_psi = tune["tirePressureRear"]
        
        min_psi = FH5_LIMITS["tire_pressure"]["min"]
        max_psi = FH5_LIMITS["tire_pressure"]["max"]
        
        if not (min_psi <= front_psi <= max_psi):
            validation_results["passed"] = False
            validation_results["errors"].append(
                f"Tire pressure front {front_psi} PSI out of range [{min_psi}, {max_psi}]"
            )
        if not (min_psi <= rear_psi <= max_psi):
            validation_results["passed"] = False
            validation_results["errors"].append(
                f"Tire pressure rear {rear_psi} PSI out of range [{min_psi}, {max_psi}]"
            )
        
        validation_results["details"]["tire_pressure"] = {
            "front": front_psi,
            "rear": rear_psi,
            "valid": (min_psi <= front_psi <= max_psi) and (min_psi <= rear_psi <= max_psi)
        }
    
    # Spring rate validation (PI class specific)
    if tune.get("springsFront") and pi_class in PI_CLASS_SPRING_RANGES:
        front_spring = tune["springsFront"]
        rear_spring = tune["springsRear"]
        
        spring_range = PI_CLASS_SPRING_RANGES[pi_class].get(tune_type, (100, 800))
        min_spring, max_spring = spring_range
        
        # Allow some tolerance beyond published ranges
        tolerance = 0.2  # 20% tolerance
        min_check = min_spring * (1 - tolerance)
        max_check = max_spring * (1 + tolerance)
        
        if not (min_check <= front_spring <= max_check):
            validation_results["errors"].append(
                f"Front spring {front_spring} LB/IN outside range [{min_spring}, {max_spring}] for {pi_class} {tune_type}"
            )
            validation_results["passed"] = False
        if not (min_check <= rear_spring <= max_check):
            validation_results["errors"].append(
                f"Rear spring {rear_spring} LB/IN outside range [{min_spring}, {max_spring}] for {pi_class} {tune_type}"
            )
            validation_results["passed"] = False
        
        validation_results["details"]["springs"] = {
            "front": front_spring,
            "rear": rear_spring,
            "valid": (min_check <= front_spring <= max_check) and (min_check <= rear_spring <= max_check),
            "range": spring_range
        }
    
    # Ride height validation
    if tune.get("rideHeightFront"):
        front_rh = tune["rideHeightFront"]
        rear_rh = tune["rideHeightRear"]
        
        min_rh = FH5_LIMITS["ride_height"]["min"]
        max_rh = FH5_LIMITS["ride_height"]["max"]
        
        if not (min_rh <= front_rh <= max_rh):
            validation_results["errors"].append(
                f"Ride height front {front_rh} IN out of range [{min_rh}, {max_rh}]"
            )
            validation_results["passed"] = False
        if not (min_rh <= rear_rh <= max_rh):
            validation_results["errors"].append(
                f"Ride height rear {rear_rh} IN out of range [{min_rh}, {max_rh}]"
            )
            validation_results["passed"] = False
        
        validation_results["details"]["ride_height"] = {
            "front": front_rh,
            "rear": rear_rh,
            "valid": (min_rh <= front_rh <= max_rh) and (min_rh <= rear_rh <= max_rh)
        }
    
    # Camber validation
    if tune.get("camberFront") is not None:
        camber_f = tune["camberFront"]
        camber_r = tune["camberRear"]
        
        min_cam = FH5_LIMITS["camber"]["min"]
        max_cam = FH5_LIMITS["camber"]["max"]
        
        if not (min_cam <= camber_f <= max_cam):
            validation_results["warnings"].append(
                f"Front camber {camber_f}Â° outside typical range [{min_cam}, {max_cam}]"
            )
        if not (min_cam <= camber_r <= max_cam):
            validation_results["warnings"].append(
                f"Rear camber {camber_r}Â° outside typical range [{min_cam}, {max_cam}]"
            )
        
        validation_results["details"]["camber"] = {
            "front": camber_f,
            "rear": camber_r,
            "valid": (min_cam <= camber_f <= max_cam) and (min_cam <= camber_r <= max_cam)
        }
    
    # Anti-roll bar validation
    if tune.get("arbFront") is not None:
        arb_f = tune["arbFront"]
        arb_r = tune["arbRear"]
        
        min_arb = FH5_LIMITS["arb"]["min"]
        max_arb = FH5_LIMITS["arb"]["max"]
        
        if not (min_arb <= arb_f <= max_arb):
            validation_results["errors"].append(
                f"Front ARB {arb_f} outside range [{min_arb}, {max_arb}]"
            )
            validation_results["passed"] = False
        if not (min_arb <= arb_r <= max_arb):
            validation_results["errors"].append(
                f"Rear ARB {arb_r} outside range [{min_arb}, {max_arb}]"
            )
            validation_results["passed"] = False
        
        validation_results["details"]["arb"] = {
            "front": arb_f,
            "rear": arb_r,
            "valid": (min_arb <= arb_f <= max_arb) and (min_arb <= arb_r <= max_arb)
        }
    
    # Final drive validation
    if tune.get("finalDrive"):
        fd = tune["finalDrive"]
        min_fd = FH5_LIMITS["final_drive"]["min"]
        max_fd = FH5_LIMITS["final_drive"]["max"]
        
        if not (min_fd <= fd <= max_fd):
            validation_results["errors"].append(
                f"Final drive {fd} outside range [{min_fd}, {max_fd}]"
            )
            validation_results["passed"] = False
        
        validation_results["details"]["final_drive"] = {
            "value": fd,
            "valid": (min_fd <= fd <= max_fd)
        }
    
    # Brake validation
    if tune.get("brakePressure") is not None:
        brake_p = tune["brakePressure"]
        brake_b = tune["brakeBalance"]
        
        min_bp = FH5_LIMITS["brake_pressure"]["min"]
        max_bp = FH5_LIMITS["brake_pressure"]["max"]
        
        if not (min_bp <= brake_p <= max_bp):
            validation_results["errors"].append(
                f"Brake pressure {brake_p} outside range [{min_bp}, {max_bp}]"
            )
            validation_results["passed"] = False
        
        validation_results["details"]["brakes"] = {
            "pressure": brake_p,
            "balance": brake_b,
            "valid": (min_bp <= brake_p <= max_bp) and (0 <= brake_b <= 100)
        }
    
    # Differential validation
    if tune.get("diffAccelRear") is not None:
        diff_accel = tune["diffAccelRear"]
        diff_decel = tune["diffDecelRear"]
        
        if not (0 <= diff_accel <= 100):
            validation_results["errors"].append(
                f"Rear diff accel {diff_accel}% outside range [0, 100]"
            )
            validation_results["passed"] = False
        
        if not (0 <= diff_decel <= 100):
            validation_results["errors"].append(
                f"Rear diff decel {diff_decel}% outside range [0, 100]"
            )
            validation_results["passed"] = False
        
        validation_results["details"]["differential"] = {
            "accel": diff_accel,
            "decel": diff_decel,
            "valid": (0 <= diff_accel <= 100) and (0 <= diff_decel <= 100)
        }
    
    # Consistency checks (warnings, not errors)
    # Front springs should generally be similar to rear (Â±30% acceptable)
    if tune.get("springsFront") and tune.get("springsRear"):
        front = tune["springsFront"]
        rear = tune["springsRear"]
        ratio = abs(front - rear) / max(front, rear)
        
        if ratio > 0.5 and tune_type not in ["drag", "drift"]:
            validation_results["warnings"].append(
                f"Large spring rate difference: Front {front} vs Rear {rear} ({ratio*100:.1f}% diff)"
            )
    
    return validation_results

print("âœ“ Validation functions defined")
```
